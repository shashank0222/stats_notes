# check normal distribution using QQ plot - Quantile Quantile plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pylab
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
def plot_data(df,feature):
    plt.figure(figsize = (10,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature] , kde=True)
    
    plt.subplot(1,2,2)
    stats.probplot(df[feature] , dist = "norm" , plot = pylab)
    
    plt.show()
```

```python
plot_data(df,"sepal_width")  # if most of the points fall under the straight line ,then it is normal distribution

```

```python
plot_data(df,"petal_width")  # not a normal distribution
```
